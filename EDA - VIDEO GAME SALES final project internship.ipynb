```python
import pandas as pd
# Load the video game sales data
data = pd.read_csv(r"C:\Users\gitan\Downloads\vgsales.csv")
```

```python
data = data.dropna(subset=['Year'])
```

```python
# Clean column names: lowercase, remove spaces
data.columns = data.columns.str.strip().str.replace(" ", "_").str.lower()

# Check again
print(data.columns.tolist())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
top_10 = data.sort_values(by='global_sales', ascending=False).head(10)

```

```python
import matplotlib.pyplot as plt
plt.title("Test Plot")
plt.show()
```

```python
genre_sales = data.groupby('genre')['global_sales'].sum().sort_values()
genre_sales.plot(kind='barh', color='teal')
plt.title("Total Global Sales by Genre")
plt.xlabel("Sales (Millions)")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
platform_sales = data.groupby('platform')['global_sales'].sum().sort_values(ascending=False).head(10)
platform_sales.plot(kind='bar', color='salmon')
plt.title("Top 10 Platforms by Global Sales")
plt.xlabel("Platform")
plt.ylabel("Global Sales (Millions)")
plt.tight_layout()
plt.show()
```

```python
games_per_year = data['year'].value_counts().sort_index()
games_per_year.plot(kind='line', marker='o')
plt.title("Games Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Games")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
publisher_sales = data.groupby('publisher')['global_sales'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=publisher_sales.values, y=publisher_sales.index, palette='magma')
plt.title("Top 10 Publishers by Global Sales")
plt.xlabel("Sales (Millions)")
plt.ylabel("Publisher")
plt.tight_layout()
plt.show()
```

```python
sales_corr = data[['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'global_sales']].corr()
sns.heatmap(sales_corr, annot=True, cmap='coolwarm')
plt.title("Correlation Between Regional Sales")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='genre', y='global_sales', data=data)
plt.xticks(rotation=45)
plt.title("Distribution of Global Sales by Genre")
plt.tight_layout()
plt.show()
```

```python
yearly_sales = data.groupby('year')['global_sales'].sum()
yearly_sales.plot(kind='line', color='green', marker='o')
plt.title("Total Global Sales by Year")
plt.xlabel("Year")
plt.ylabel("Global Sales (Millions)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
heatmap_data = data.pivot_table(index='platform', columns='year', values='name', aggfunc='count', fill_value=0)
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5)
plt.title("Number of Games Released by Platform and Year")
plt.tight_layout()
plt.show()
```

```python
game_name = "Wii Sports"

# Filter and check
if game_name in data['name'].values:
    game_data = data[data['name'] == game_name]
    
    print(game_data[['name', 'platform', 'year', 'genre', 'publisher', 
                     'na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'global_sales']])
    
    # Plot regional sales
    sales = game_data[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].iloc[0]
    sns.barplot(x=sales.index, y=sales.values, palette='Set2')
    plt.title(f"Regional Sales Breakdown for '{game_name}'")
    plt.ylabel("Sales (Millions)")
    plt.xlabel("Region")
    plt.tight_layout()
    plt.show()
else:
    print(f" Game '{game_name}' not found in the dataset.")
```

```python
import pandas as pd

# Load your dataset
data = pd.read_csv(r'C:\Users\gitan\Downloads\vgsales.csv')

# Optional: Clean column names for consistency
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')

```

```python
genre_trend = data.groupby(['year', 'genre'])['global_sales'].sum().unstack()
genre_trend.plot(kind='line', figsize=(12, 6))
plt.title("Global Sales by Genre Over Years")
plt.xlabel("Year")
plt.ylabel("Global Sales (Millions)")
plt.legend(title='Genre')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_publishers = data.groupby('publisher')['global_sales'].sum().sort_values(ascending=False).head(5)
top_publishers.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), title='Top Publishers by Global Sales')
```
